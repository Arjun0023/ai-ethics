```python
# Lab 1 :-Exploratory Data Analysis on 2 Variable Linear Regression Model.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
df = pd.read_csv('/content/Iris.csv')

print("First few rows of the dataset :\n", df.head())
```

```python
x = df['PetalLengthCm']
y = df['PetalWidthCm']
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c=x, cmap='Oranges_r')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('scatter Plot :Petal Length vs. Petal Width')
plt.show()
```

```python
correlation = np.corrcoef(x, y)[0, 1]
print(f"Correlation coefficient between Petal Length and Petal Width: {correlation}")
```

```python
model = LinearRegression()
x_reshaped = x.values.reshape(-1, 1)
model.fit(x_reshaped, y)
```

```python
y_pred = model.predict(x_reshaped)
plt.figure(figsize=(8, 6))
plt.scatter(x, y,color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Linear Regression :Petal Length vs. Petal Width')
plt.legend()
plt.show()
```

```python
print(f"slope (coefficient): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```
