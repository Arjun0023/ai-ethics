```python
import gensim.downloader as api
import numpy as np
model = api.load('word2vec-google-news-300')
word_a, word_b , word_c = 'rupee', 'india','dollar'
result = model.most_similar(positive=[word_b, word_c], negative=[word_a],
topn=1)
word_d = result[0][0]
print(f"'{word_a}' is to '{word_b}' as '{word_c}' is to '{word_d}'")
print(f"Word: {word_d}, Similarity: {result[0][1]}")
words = [word_a ,word_b, word_c, word_d]
word_vsctors = np.array([model[word] for word in words])
for i, word in enumerate(words):
  print(f" Vector for '{word}' : {word_vsctors[i]}" )
```

```python
paragraph="""
Artificial Intelligence And Machine Learning are revolutionizing
industries across the world.
Technology is growing at an unprecedented rate ,making innovation a
crucial factor for businesses to stay competitive.
"""
words = paragraph.lower().split()
key_terms = ['artificial', 'intelligence', 'technology', 'innovation']
for term in key_terms:
  if term in model:
    similar_words = model.most_similar(term, topn=5)
    print(f"Words similar to '{term}':")
    for word, similarity in similar_words:
      print(f"{word}: {similarity}")
      print("\n")
  else:
    print(f"'{term}' not found in the model's vocabulary.\n")
```
